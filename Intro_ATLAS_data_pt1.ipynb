```python
import uproot
import pandas as pd
import matplotlib.pyplot as plt
```

# Lab 5: Intro to ATLAS Data, Part 1 - Kyle Pitz

```python
data = uproot.open("https://atlas-opendata.web.cern.ch/atlas-opendata/samples/2020/1largeRjet1lep/MC/mc_361106.Zee.1largeRjet1lep.root")[b'mini;1']
```

```python
lep_n = data.pandas.df(['lep_n'])
```

```python
fig = plt.hist(lep_n.loc[:,'lep_n'], bins = 5, range = (0,5))
title = plt.title('Number of Leptons')
xlabel = plt.xlabel('# of leptons')
ylabel = plt.ylabel('# of entries')
```

```python
fig2 = plt.hist(lep_n.loc[:,'lep_n'], bins = 5, range = (0,5), density = True)
title = plt.title('Number of Leptons, Normalized')
xlabel = plt.xlabel('# of leptons')
ylabel = plt.ylabel('# of entries')
```
